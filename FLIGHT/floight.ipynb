```python
## feature engineering
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
## drop the Date of Journey
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_Hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df['Arrival_Hour']=df['Arrival_Hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df.info()
```

```python
df.head(2)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['Dep_hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_min']=df['Dep_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_min']=df['Dep_min'].astype(int)
```

```python
df.info()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
df['Duration_hour'] = df['Duration'].str.extract(r'(\d+)h').fillna(0)
df['Duration_hour'] = df['Duration_hour'].astype(int)
```

```python
df['Duration_min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df['Duration_min'].unique()
```

```python
df['Duration_min'].mode()
```

```python
df['Duration_min']=pd.to_numeric(df['Duration_min'],errors='coerce')
```

```python
df.head(2)
```

```python
mode_value = df['Duration_min'].mode()[0]
df['Duration_min'].fillna(mode_value , inplace=True)
```

```python
df['Duration_min']=df['Duration_min'].astype(int)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
```

```python
encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
```

```python
encoded_df=pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
df.reset_index(drop=True , inplace=True)
encoded_df.reset_index(drop=True,inplace=True)
```

```python
df = df.drop(['Airline','Source','Destination'],axis=1)
```

```python
df = pd.concat([df , encoded_df],axis=1)
```
